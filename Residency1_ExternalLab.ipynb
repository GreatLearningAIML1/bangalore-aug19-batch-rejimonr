## About the dataset

This is perhaps the best known database to be found in the pattern recognition literature. The data set contains 3 classes of 50 instances each, where each class refers to a type of iris plant.  One class is linearly separable from the other 2; the latter are NOT linearly separable from each other.

Attribute Information:
   1. sepal length in cm
   2. sepal width in cm
   3. petal length in cm
   4. petal width in cm
   5. class: 
      -- Iris Setosa
      -- Iris Versicolour
      -- Iris Virginica

### Multiply Sepal Length and width and store it under the column name "SepalExtra" in the same Iris dataframe

```python
#New derived column SepalExtra based on product of Sepal Lenght and Width
iris["SepalExtra"] = iris["Sepal Length (in cm)"]*iris["Sepal Width in (cm)"]
```

```python
#Check if the new column got added
iris.head()
```

### Find out the mean and variance for each column but for class column 

```python
#Mean and Variance using inbuilt methods
print("Mean of all numeric columns\n")
print(iris.mean())
print("\n")
print("Variance of all numeric columns\n")
print(iris.var())
```

### Find the range of all the columns in the dataset

*Range = Max value - Min value (in the column)*

```python
#We will use the inbuilt functions max and min on numeric columns and then use the standard formula of range
#We will take advantage of the series operations to do this in one statment

print("Range of all numeric columns\n")
print(iris.max(numeric_only=True) - iris.min(numeric_only=True))
```

### Sort the entire dataset according to the column Petal width

### Remove the new column "SepalExtra" from the dataframe

```python
#Remove derived columnn SepalExtra using inbuilt function drop and make inplace True to make it permanent
print("Dropping column SepalExtra\n")
iris.drop(columns=["SepalExtra"],inplace=True)
print("Sample to see no derived column SepalExtra\n")
print(iris.head())
```
